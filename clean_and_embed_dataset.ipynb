```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("resume_text.csv")
data = data.drop('Unnamed: 0', axis=1)
data['fontname'] = data['fontname'].astype('category')
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
def word_to_vector(word):
    token = nlp(word)
    if token.has_vector:
        return token.vector
    else:
        # fallback for unknown words
        return np.zeros(300)
```

```python
data['word'] = data['word'].apply(lambda x: word_to_vector(x))
```

```python
embedding_matrix = np.vstack(data['word'].values)
embedding_matrix
```

```python
embedding_df = pd.DataFrame(embedding_matrix, index=data.index)
embedding_df.columns = [f"emb_{i}" for i in range(embedding_df.shape[1])]
embedding_df
```

```python
meta_df = data.drop(columns=['word'])
data_emb = pd.concat([embedding_df, meta_df], axis=1)
data_emb
```

```python
from category_encoders import OrdinalEncoder
import joblib
```

```python
encoder = OrdinalEncoder()
encoder.fit(data_emb['fontname'])
```

```python
data_emb['fontname'] = encoder.transform(data_emb['fontname'])
encoder = joblib.dump(encoder, 'ordinal_encoder.pkl')
```

```python
data_final = data_emb
data_final.to_csv('dataset.csv')
```
