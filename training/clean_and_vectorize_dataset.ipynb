```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv("resume_text.csv")
data_1 = data_1.drop('Unnamed: 0', axis=1)
data_1['fontname'] = data_1['fontname'].astype('category')
```

```python
data_2 = pd.read_csv("new_resumes_to_add.csv")
data_2 = data_2.drop('Unnamed: 0', axis=1)
data_2['fontname'] = data_2['fontname'].astype('category')
```

```python
data_2["label"] = data_2["label"].replace("0", np.nan)
```

```python
data_2["label"].value_counts()
```

```python
data_2
```

```python
data = pd.concat([data_1, data_2], axis=0)
```

```python
data["fontname"].value_counts()
```

```python
data.dtypes
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
def word_to_vector(word):
    token = nlp(word)
    if token.has_vector:
        return token.vector
    else:
        # fallback for unknown words
        return np.zeros(300)
```

```python
data['word'] = data['word'].apply(lambda x: word_to_vector(x))
```

```python
vectorized_matrix = np.vstack(data['word'].values)
vectorized_matrix
```

```python
vectorized_df = pd.DataFrame(vectorized_matrix, index=data.index)
vectorized_df.columns = [f"vec_{i}" for i in range(vectorized_df.shape[1])]
vectorized_df
```

```python
meta_df = data.drop(columns=['word'])
data_vec = pd.concat([vectorized_df, meta_df], axis=1)
data_vec
```

```python
from category_encoders import OrdinalEncoder
import joblib
```

```python
encoder = OrdinalEncoder()

encoder.fit(data['fontname'])
```

```python
data_vec['fontname'] = encoder.transform(data['fontname'])

encoder = joblib.dump(encoder, '../models/ordinal_encoder.pkl')
```

```python
encoder = OrdinalEncoder()

encoder.fit(data_vec[['label']])
```

```python
data_vec['label'] = encoder.transform(data_vec['label'])

encoder = joblib.dump(encoder, '../models/resume_label_encoder.pkl')
```
