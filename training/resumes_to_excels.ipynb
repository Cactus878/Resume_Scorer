```python
import pandas as pd
import pdfplumber
import os
```

```python
dir = "../new_resumes"
data = []

for file in os.listdir(dir):
    filename = os.fsdecode(file)
    if filename.endswith(".pdf"):
        with pdfplumber.open(f'{dir}/{filename}') as pdf:
            for page in pdf.pages:
                words = page.extract_words(extra_attrs=["fontname", "size"])
                for w in words:
                    data.append({"word": w["text"],
                                "width": w["width"],
                                "height": w["height"],
                                "left_distance": w["x0"], 
                                "bottom_distance": w["bottom"], 
                                "right_distance": w["x1"], 
                                "top_distance": w["top"],
                                "fontname": w["fontname"],
                                "fontsize": w["size"],
                                "label": 0})


df = pd.DataFrame(data)
```
